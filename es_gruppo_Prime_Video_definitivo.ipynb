```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_df = pd.read_csv('dataset.csv')
spotify_df.head()
```

```python
spotify_df.info()
```

```python
spotify_subset = spotify_df[['artists', 'album_name', 'track_name', 'popularity', 
                            'duration_ms','danceability','energy','loudness',
                            'speechiness','acousticness','explicit' ,'instrumentalness','liveness',
                            'valence','tempo','track_genre']]
spotify_subset.head()
```

```python
plt.style.use('fivethirtyeight')

popolarita_genere = spotify_subset.groupby(['track_genre'], as_index = False)['popularity'].mean().sort_values('popularity', ascending = False)

popolarita_genere
p = popolarita_genere.head(10)
p    
```

```python
popolarita_genere.track_genre.unique(), len(popolarita_genere.track_genre.unique())
```

```python
plt.figure(figsize = (15, 7))
plt.bar(data = p, x = 'track_genre', height = 'popularity')

plt.title('Top 10 generi per popolaritÃ ')
plt.xlabel('Genere')
plt.ylabel('PopolaritÃ ')
plt.ylim(40, 65);
```

```python
durata_media_ms = spotify_subset['duration_ms'].mean()
num_tracce_inferiori = (spotify_subset['duration_ms'] < durata_media_ms).sum()
num_tracce_superiori = (spotify_subset['duration_ms'] >= durata_media_ms).sum()
percentuale_inferiore = (num_tracce_inferiori / len(spotify_subset)) * 100
percentuale_superiore = (num_tracce_superiori / len(spotify_subset)) * 100
labels = ['Durata inferiore', 'Durata superiore']
sizes = [percentuale_inferiore, percentuale_superiore]
plt.pie(sizes, labels = labels, autopct='%1.1f%%')
plt.legend(labels=[f'Durata inferiore: {num_tracce_inferiori} ', f'Durata superiore: {num_tracce_superiori} '],bbox_to_anchor = (1.01, 1));
plt.title(f'Percentuale di tracce con durata inferiore e superiore alla media {round(durata_media_ms,2)}')
```
